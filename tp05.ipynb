```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
space_img = cv2.imread('/content/Space.bmp', cv2.IMREAD_GRAYSCALE)
gappedtext_img = cv2.imread('/content/GappedText.bmp', cv2.IMREAD_GRAYSCALE)
joker_img = cv2.imread('/content/Joker.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
def display_images(images, titles, cmap='gray'):
    plt.figure(figsize=(12, 6))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(img, cmap=cmap)
        plt.title(title)
        plt.axis('off')
    plt.show()


mean_15x15 = cv2.blur(space_img, (15, 15))
_, space_bw = cv2.threshold(mean_15x15, 127, 255, cv2.THRESH_BINARY)

display_images([space_img, mean_15x15, space_bw],
               ["Original Space.bmp", "Mean Filter 15x15", "Black & White"])
```

```python
mean_3x3 = cv2.blur(gappedtext_img, (3, 3))
_, gappedtext_bw = cv2.threshold(mean_3x3, 127, 255, cv2.THRESH_BINARY)

display_images([gappedtext_img, mean_3x3, gappedtext_bw],
               ["Original GappedText.bmp", "Mean Filter 3x3", "Black & White"])
```

```python
mean_5x5 = cv2.blur(joker_img, (5, 5))
mean_15x15 = cv2.blur(joker_img, (15, 15))
mean_25x25 = cv2.blur(joker_img, (25, 25))
mean_29x29 = cv2.blur(joker_img, (29, 29))

display_images([joker_img, mean_5x5, mean_15x15, mean_25x25, mean_29x29],
               ["Original Joker.bmp", "Mean 5x5", "Mean 15x15", "Mean 25x25", "Mean 29x29"])
```

```python
gaussian_5x5 = cv2.GaussianBlur(joker_img, (5, 5), 0)

display_images([mean_5x5, gaussian_5x5],
               ["Mean Filter 5x5", "Gaussian Filter 5x5"])
```

```python
gaussian_15x15 = cv2.GaussianBlur(joker_img, (15, 15), 0)

display_images([mean_15x15, gaussian_15x15],
               ["Mean Filter 15x15", "Gaussian Filter 15x15"])
```
